## The pipeline from movement to awake-asleep classification

Most, if not all, methods of automatic sleep scoring in use in smart watches today is based on triaxial accelerometer data. This applies to both clinical sleep tracking and smart devices in tech. These data record the acceleration in the x-, y-, and z-axis of a smart watch or phone, measured in units of the Earth's gravity _g_ = 9.8 m/s^2; typically, sampling is at 50 or 100 Hz.  While there are additionally models developed to use PPG, there are issues related to skin tone and reflectivity that limits the generalizeability of such models.

We'll use the Pisces library to analyze how the performance of motion-based sleep scoring algorithms is impacted by the addition of environmental noise. The first step is to point the `DataSetObject` class at the directory containing our training + test data, which returns a dictionary like

```{ ..., "folder_name": DataSetObject, ...}```

It is now easy to plot the accelerometer, etc., stored in our data set folders.

````python
# Let's load some accelerometer data and plot it
from pisces.data_sets import DataSetObject

import os
from pathlib import Path

CWD = Path(os.getcwd())
# hardcoded data locations are easiest, if most brittle
data_location = CWD.parent.parent.joinpath('data')
print("Looking for data in ", data_location)

# This assumes you have set up the walch_et_al data set. See [this notebook]() for how to do that.
sets = DataSetObject.find_data_sets(data_location)
walch = sets['walch_et_al']
walch.parse_data_sets()
print(f"Found {len(walch.ids)} subjects")
````

````python
walch_0 = walch.ids[0]

w0_accel = walch.get_feature_data( 'accelerometer', walch_0,).to_numpy()
````

````python
w0_accel.shape
````

````python
from matplotlib import pyplot as plt
import seaborn as sns

PALETTE = sns.color_palette('colorblind')
# Setting font and style parameters
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = 18

def set_plot_style(ax):
    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Customize the remaining spines (left and bottom)
    ax.spines['left'].set_linewidth(1.2)
    ax.spines['bottom'].set_linewidth(1.2)

def plot_accel_data(ax, w0_accel):
    axis_names = ['Time (h)', 'X-axis', 'Y-axis', 'Z-axis']

    w0_time = w0_accel[:, 0] / 3600

    for i in range(1, 4):
        ax.plot(w0_time, w0_accel[:, i], label=axis_names[i], color=PALETTE[i+1])

    # draw horizontal -- lines at +/- 1g
    ax.axhline(1, color='black', linestyle='--')
    ax.axhline(-1, color='black', linestyle='--')

    ax.set_xlim(w0_time[0], w0_time[-1])
    ax.set_title("Accelerometer data")
    ax.set_xlabel(axis_names[0])
    ax.set_ylabel("Acceleration (g)")
    ax.legend()
    set_plot_style(ax)

fig, ax = plt.subplots(1, 1, figsize=(20, 10))
plot_accel_data(ax, w0_accel)
````

For clinical sleep tracking, the standard has long been Cole-Kripke like methods based on highly filtered and aggregated _actigraphy_ data. The specifics of the filtering and aggregation were proprietary until a few years ago (e.g. ActiGraph algorithm on GitHub) though it is possible to reverse-engineer these algorithms on specific data sets (e.g. te Lindert, _et al_). Once processed, a weighted sum of this data in a sliding window is taken, with weights determined by gradient descent methods.

Pisces has Arcascope's ADS processing, which in our experiments we have found is more robust to noise (more on this later). The library also has _te Lindert et al_'s reverse-engineered algorithm. The `agcounts` library could be incorporated, except it requires python < 3.10 and pisces requires 3.11+.


````python
from pisces.utils import build_ADS, build_activity_counts_te_Lindert_et_al

# Let's build some activity counts

# Arcascope's noise-robust activity detection system
w0_ads = build_ADS(w0_accel) 
# te Lindert et al's activity detection system
# There are filters that calibrate on the inputs, hence added noise leads to filtering of information
w0_tL = build_activity_counts_te_Lindert_et_al(w0_accel) 


fig, ax = plt.subplots(1, 1, figsize=(10, 5))

ax.plot(w0_tL[0], w0_tL[1], label='te Lindert et al')
ax.plot(w0_ads[0], w0_ads[1], label='ADS')
ax.legend()
set_plot_style(ax)
````

The field of automatic scoring in consumer tech smart watches has clustered around more sophisticated techniques utilizing the high-resolution and high fidelity of consumer accelerometers. One well-known approach is to take a spectrogram, or power-spectrum decomposition, of this data. See, e.g., (Canton & Walch GitHub), Mads Olsen, ...

Spectrograms are processed by deep neural networks akin to image segmenters, which instead of outputting _a cat is over here_ or _a dump truck is here_ identifies _this segment looks like someone in REM sleep_, etc.

````python
# We provide an example of a spectrogram. 
# Specific ML models will have their own configurations of NFFT, pre-transform resampling, etc.

import numpy as np


w0_time = w0_accel[:, 0]
w0_accel_x = w0_accel[:, 1]
w0_accel_y = w0_accel[:, 2]
w0_accel_z = w0_accel[:, 3]
w0_accel_amp = np.sqrt(w0_accel_x**2 + w0_accel_y**2 + w0_accel_z**2)

# Compute the median sampling rate
median_Hz = int(1 / np.median(np.diff(w0_time)))
nfft = 512
overlap = nfft // 2


# Compute the spectrogram for each axis
plt.figure(figsize=(10, 5))
spec, _, _, _ = plt.specgram(w0_accel_amp, NFFT=nfft, Fs=median_Hz, noverlap=overlap, cmap='viridis')
plt.colorbar(label='Intensity (dB)')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Spectrogram of Accelerometer Data')
````

## The challenge of Noise

Accelerometers are highly sensitive instruments, and record buzzing in their environments. When a smart watch is strapped to a human in such an environment, some of the buzzing is damped but the rest is recorded by the device's accelerometer, masking movements of the subject. Said differently, the added white noise changes the statistics of the signal, so one cannot expect machine learning models to be have the same as when they are trained and evaluated on accelerometer data that does not have the noise.

We have a data set, `hybrid_motion`, which is a copy of the `walch_et_al` data set except the accelerometer has been acted on by gyroscopic measurements from a US Navy vessel to make a "hybrid" of wrist-worn accelerometer and ship-board motion.

If you do not have access to this data set, you can instead take a copy of `walch_et_al` replace the accelerometer CSVs after adding `np.random.normal` values with various scalings and shapes. Save this data in `data/hybrid_motion` along side `data/walch_et_al`. See `./noise_study.ipynb` for an example of increasing the noise even further in our hybrid motion  data set. 

````python

hybrid = sets['hybrid_motion']
hybrid.parse_data_sets()
print(f"Found {len(hybrid.ids)} subjects")
hybrid_0 = walch_0  # same data set ids, accelerometer data has just been modified

h0_accel = hybrid.get_feature_data( 'accelerometer', hybrid_0,).to_numpy()
````

````python
def hybrid_overlay(ax, h0_accel, w0_accel):
    axis_names = ['Time (h)', 'X-axis', 'Y-axis', 'Z-axis']
    w0_time = w0_accel[:, 0] / 3600
    h0_time = h0_accel[:, 0] / 3600

    for i in range(1, 4):
        ax.plot(h0_time, h0_accel[:, i], label=axis_names[i], color=PALETTE[i+1])
        # ax.plot(w0_time, w0_accel[:, i], c='black', alpha=0.8)

    # draw horizontal -- lines at +/- 1g
    ax.axhline(1, color='black', linestyle='--')
    ax.axhline(-1, color='black', linestyle='--')

    ax.set_xlim(h0_time[0], h0_time[-1])
    ax.set_title("Noisy 'Hybrid' Accelerometer data")
    ax.set_xlabel(axis_names[0])
    ax.set_ylabel("Acceleration (g)")
    ax.legend()
    set_plot_style(ax)
````

````python

fig, ax = plt.subplots(1, 1, figsize=(20, 10))
hybrid_overlay(ax, h0_accel, w0_accel)
````

# Evaluation techniques

The PSG recordings we consider here are spaced 30 seconds apart, which we call the "epoch length" (not to be confused with training epochs).  The data set has highly imbalanced classes, since the PSG recordings are of sleeping subjects. A trivial ML model that outputs "Sleep" for every epoch automatically scores 80+% accuracy, simply because most subjecst slept at least 7 of the 8ish hours they were under recording. The result is that simply looking at accuracy (\% of all PSG labels classified correctly) is not a good way to evaluate the performance of our models.

Furthermore, for our applications we are much more concerned with missing wake epochs at high sleep accuracies.

We will use the following metrics to evaluate the performance of our models:
1. Area under the ROC curve (AUROC) - this is a measure of how well the model can distinguish between sleep and wake epochs.
2. WASA95 - this is the percentage of wake epochs that are misclassified as sleep epochs, when the model is tuned to have 95% sleep accuracy. This is also known as the specificity, when sleep is the positive binary class and we choose thresholds in regression outputs to get 95% sensitivity.

# ROC Explainer

ROC is short for Receiver Operating Characteristic. It is a plot of the true positive rate (sensitivity) against the false positive rate (1-specificity) for the different possible thresholds of a binary classifier. The area under the ROC curve (AUROC) is a measure of how well the model can distinguish between the two classes. An AUROC of 0.5 means the model is no better than random, while an AUROC of 1 means the model is perfect.

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def roc_explainer(ax):
    # Color palette
    color_palette = sns.color_palette("colorblind")

    # Data for the curve
    DECAY = 9
    def roc_map(x):
        return 1 - np.exp(-x * DECAY)
    def roc_inv_map(y):
        return -np.log(1 - y) / DECAY 
    x = np.linspace(0, 1, 100)
    y = roc_map(x)

    # WASA80 and WASA90 points
    wasa80_y = 0.8
    wasa80_x = roc_inv_map(wasa80_y)
    wasa90_y = 0.9
    wasa90_x = roc_inv_map(wasa90_y)

    # Plotting
    # fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(x, y, color=color_palette[2], linewidth=2)

    # Adding points for WASA80 and WASA90
    ax.plot(wasa80_x, wasa80_y, 'o', color=color_palette[0], markersize=8, label="WASA80")
    ax.plot(wasa90_x, wasa90_y, 'o', color=color_palette[1], markersize=8, label="WASA90")

    # Adding dashed lines for visual reference
    ax.hlines(wasa80_y, 0, wasa80_x, color='black', linestyle='--', linewidth=1)
    ax.vlines(wasa80_x, 0, wasa80_y, color='black', linestyle='--', linewidth=1)
    ax.hlines(wasa90_y, 0, wasa90_x, color='black', linestyle='--', linewidth=1)
    ax.vlines(wasa90_x, 0, wasa90_y, color='black', linestyle='--', linewidth=1)

    # Annotations for the WASA points using f-strings and aligned vertically at x=0.6
    text_x = 0.6  # Fixed x position for annotations
    text_wasa80_y = 0.4  # Custom y position for WASA80 annotation
    text_wasa90_y = text_wasa80_y + 0.2 # Custom y position for WASA90 annotation
    arrow_offset_y = 0.05
    arrow_offset_x = 0.01

    # Adding WASA80 annotation with an arrow
    ax.text(
        text_x, text_wasa80_y,
        f"1 - WASA80 = {1 - wasa80_x:.2f}\nWASA80 = {wasa80_x:.2f}", 
        color=color_palette[0], ha='center', va='center'
    )
    ax.annotate(
        '', xy=(wasa80_x + arrow_offset_x, wasa80_y), xytext=(text_x, text_wasa80_y + arrow_offset_y),
        arrowprops=dict(arrowstyle="->", color=color_palette[0])
    )

    # Adding WASA90 annotation with an arrow
    ax.text(
        text_x, text_wasa90_y,
        f"1 - WASA90 = {1 - wasa90_x:.2f}\nWASA90 = {wasa90_x:.2f}", 
        color=color_palette[1], ha='center', va='center'
    )
    ax.annotate(
        '', xy=(wasa90_x + arrow_offset_x, wasa90_y), xytext=(text_x, text_wasa90_y + arrow_offset_y),
        arrowprops=dict(arrowstyle="->", color=color_palette[1])
    )

    # Axis labels
    ax.set_xlabel("Fraction of wake scored as sleep")
    ax.set_ylabel("Fraction of sleep scored as sleep")


    # Set limits
    ax.set_xlim(0, 1.0)
    ax.set_ylim(0, 1.1)

    set_plot_style(ax)

_, ax = plt.subplots(figsize=(8, 6))
roc_explainer(ax)
# Show plot
plt.tight_layout()
plt.show()
````

# LOOX

For each subject in the original Walch study, we have created 2 separate accelerometer data sets by adding noise to the original. Thus, when using the common leave-one-out cross-validation technique, we can exchange the test subject for the noisy version of the test subject. This allows us to evaluate the performance of our models on unseen data whose distribution is similar to the training set, as well as noisy data. We call this technique LOOX, for Leave-One-Out with eXchange.

````python
import matplotlib.pyplot as plt
import matplotlib.table as tbl

def add_loox_explainer(ax):
    # Data for the table
    data = [
        ["A", "B", "C", "$\\mathcal{C}$"],
        ["A", "C", "B", "$\\mathcal{B}$"],
        ["B", "C", "A", "$\\mathcal{A}$"],
    ]

    data_line_loo = f"{data[0][0]}, {data[0][1]}, {data[0][2]}"
    data_line_loox = f"{data[2][-1]}, {data[1][-1]}, {data[0][-1]}"

    # Headers for the table
    columns = [f"Train\n{data_line_loo}", f"Test - LOO\n{data_line_loo}", f"Test - LOOX\n{data_line_loox}"]

    # Create the table
    table_data = [[row[0] + ", " + row[1], row[2], row[3]] for row in data]
    table = tbl.table(
        ax,
        cellText=table_data,
        colLabels=columns,
        cellLoc="center",
        loc="center",
        bbox=[0, 0, 1, 1]  # Make the table fill the entire axis
    )

    # Set font sizes
    table.auto_set_font_size(False)
    for (i, j), cell in table.get_celld().items():
        if i == 0:
            cell.set_fontsize(24)  # Header font size
        else:
            cell.set_fontsize(30)  # Data cell font size
    # Adjust the table size
    table.scale(1, 2)
    ax.set_axis_off()

    # Change cell colors
    for (i, j), cell in table.get_celld().items():
        if i == 0:
            cell.set_facecolor("#f0f0f0")  # Header cells
        else:
            cell.set_facecolor("#ffffff")  # Data cells
            if j == 1:
                cell.set_facecolor(PALETTE[0])  # Highlight LOO column
                cell.set_alpha(0.5)  # Set opacity to 50%
            elif j == 2:
                cell.set_facecolor(PALETTE[1])  # Highlight LOOX column
                cell.set_alpha(0.5)  # Set opacity to 50%
````

````python
plt.figure(figsize=(10, 5))
add_loox_explainer(plt.gca())
````
