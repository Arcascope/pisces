```python
import numpy as np
import keras

N_CLASSES = 2
N_SAMPLES = 1024

specAt = keras.metrics.SensitivityAtSpecificity(0.9, class_id=0)
y_true = np.random.randint(0, N_CLASSES, size=(N_SAMPLES,))
y_true_one_hot = keras.utils.to_categorical(y_true, num_classes=3)  # One-hot encode y_true

y_pred = np.random.normal(loc=y_true_one_hot, scale=0.5).clip(0, 1)  # Random predictions
specAt.update_state(y_true=y_true_one_hot, y_pred=y_pred)
```

```python
specAt.result().numpy()
```

```python
from matplotlib import pyplot as plt


plt.plot(specAt.true_positives.numpy(), specAt.false_positives.numpy())
```
